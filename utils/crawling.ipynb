```python
# Reference
# 한국바텐더협회 : https://www.beveragemaster.kr/

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
df = pd.DataFrame(columns=['name', 'skill', 'glass', 'ingredients'])

headers = {'User-Agent': 'Mozilla/5.0'}

#URL = "https://www.beveragemaster.kr/license_practice"

for i in range(4):
    URL = f"https://www.beveragemaster.kr/index.php?mid=license_practice&page={i+1}"
    request = requests.get(URL, headers=headers)
    soup = BeautifulSoup(request.text, 'html.parser')
    recipe_list = soup.find_all(attrs = {"class":'board_list_document'})
    for recipe in recipe_list:
        details = recipe.find_all('dd')
        img = details[0]
        record = []
        name = details[1].a.string.strip()
        record.append([re.search('[가-힣\d\s\-]+',name).group(), re.search('[a-zA-Z\d\s\-]+',name).group()])
        record.append([x.string for x in details[2].find_all('span')])
        record.append([x.string for x in details[3].find('span')])
        record.append(details[4].text.strip('\t').split('\n'))
        df.loc[len(df)] = record
```
